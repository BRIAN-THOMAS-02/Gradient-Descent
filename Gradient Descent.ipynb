### 1) Find the Derevative of Eqaution w.r.t. m = 2/n -x(mx+b)

### 2) Find the Derevative of Eqaution w.r.t. b = 2/n -(mx+b)

```python
cost_f = []
y_pred = []
derev_m = []
derev_b = []
grad_m = []
grad_b = []


def gradient_descent(x,y):
    #m_curr, b_curr = 0.006934399691920918, 25.96434091236954
    m_curr = b_curr = 0
    iterations = 100
    n = len(x)
    learning_rate = 0.0001

    print('i |         m          |          b           |         cost       |')
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        #print(y_predicted)
        y_pred.append(y_predicted)
        
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        cost_f.append(cost)
        
        md = -(2/n)*sum(x*(y-y_predicted))
        derev_m.append(md)
        
        bd = -(2/n)*sum(y-y_predicted)
        derev_b.append(bd)
        
        m_curr = m_curr - learning_rate * md
        grad_m.append(m_curr)
        
        b_curr = b_curr - learning_rate * bd
        grad_b.append(b_curr)
        
        print(i, '|', m_curr, '  |  ', b_curr, '  |  ', cost)
```

# This is the Slope of the Cost Function f(x) = x

### Basically we find the derevative of this function, which gives us the Slope

```python
import matplotlib.pyplot as plt
from sympy.plotting import plot
import sympy as sp

x = sp.Symbol("x")
f = x**2
derivative = sp.diff(f, x)

plot(derivative, ylabel = "f'(x)")
```
